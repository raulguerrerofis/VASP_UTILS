**Plot of bands with my library bandextractor**

```python
import vsp_band_plotter.read_bands_xml  as rvasp
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
path  = '/scratch/rguerrero/TLG/Multilayers/bilayergraphene_trigonalwarping/fromrelaxation/sampling_around_K'+'/'   #path where the python code is 
file1 = path+'vasprun.xml'  #path where the vasprun is

band = rvasp.bands_df(file1).eigh(plot=True, 
                                  hsp_names = [r'K-$\delta$','K',r'K+$\delta$'], 
                                  ylim=(-0.01,0.01),title=r'$\delta$-K-+$\delta$',
                                  marker='.',
                                  dpi=350,
                                  xhsp=[0,3],xshift=1380)
```

```python
band2 = band
fig2,ax2 = plt.subplots()
sec1 = band2.loc[0:302]
sec2 = band2.loc[303:350]
# display(sec1[['band 70', 'band 170']].head(1))
sec2 = sec2.set_axis(labels = range(302,350),axis=0)

sec1.plot(legend=False,color='blue',ylim=(-1,1),lw=0.75,xlim=(250,350),ax=ax2,marker='.',markersize=2)
sec2.plot(legend=False,color='blue',ylim=(-1,1),lw=0.75,xlim=(250,350),ax=ax2,marker='.',markersize=2)
# band2.plot(legend=False,color='blue',ylim=(-0.5,0.5),lw=0.75,xlim=(250,350),ax=ax2)
ax2.axvline(302,lw=0.75,c='gray',zorder=0)
ax2.axhline(0.0,lw=0.75,c='gray',zorder=0)
ax2.set_ylabel('energy (eV)')
ax2.set_xticks([302])
ax2.set_xticklabels(['X1|X'])
ax2.set_title(r'$\theta$=$10^{\circ}$')
fig2.savefig(path+'final_band_plot.png',format='png',dpi=350)


```
